### Stacking

```python
# Load các mô hình cơ sở đã lưu
linear_model = joblib.load('../Data/linear_regression_model.pkl')
lasso_model = joblib.load('../Data/lasso_regression_model.pkl')
mlp_model = joblib.load('../Data/mlp_regression_model.pkl')

# Khởi tạo K-Fold
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Dự đoán cho từng mô hình cơ sở bằng cross-validation
train_pred_linear_cv = cross_val_predict(linear_model, X_train_scaled, y_train, cv=kf)
train_pred_lasso_cv = cross_val_predict(lasso_model, X_train_scaled, y_train, cv=kf)
train_pred_mlp_cv = cross_val_predict(mlp_model, X_train_scaled, y_train, cv=kf)

# Ghép các dự đoán thành ma trận đặc trưng cho mô hình meta
train_meta_X_cv = np.column_stack((train_pred_linear_cv, train_pred_lasso_cv, train_pred_mlp_cv))

# Khởi tạo mô hình meta - Lasso Regression
meta_model = Lasso()

# Tạo lưới tham số để tìm alpha tốt nhất
param_grid = {'alpha': np.logspace(-4, 0, 50)}
grid_search = GridSearchCV(meta_model, param_grid, cv=kf, scoring='r2')
grid_search.fit(train_meta_X_cv, y_train)

# Lấy alpha tốt nhất từ GridSearch
best_alpha = grid_search.best_params_['alpha']

# Huấn luyện lại mô hình meta với alpha tối ưu
meta_model_opt = Lasso(alpha=best_alpha)
meta_model_opt.fit(train_meta_X_cv, y_train)

# Dự đoán trên tập train
train_meta_pred = meta_model_opt.predict(train_meta_X_cv)
r2_train, mse_train, rmse_train, mae_train = evaluate_model(y_train, train_meta_pred)

# Dự đoán trên tập validation
val_pred_linear = linear_model.predict(X_validation_scaled)
val_pred_lasso = lasso_model.predict(X_validation_scaled)
val_pred_mlp = mlp_model.predict(X_validation_scaled)
val_meta_X = np.column_stack((val_pred_linear, val_pred_lasso, val_pred_mlp))
val_meta_pred = meta_model_opt.predict(val_meta_X)
r2_val, mse_val, rmse_val, mae_val = evaluate_model(y_validation, val_meta_pred)

# Dự đoán trên tập test
test_pred_linear = linear_model.predict(X_test_scaled)
test_pred_lasso = lasso_model.predict(X_test_scaled)
test_pred_mlp = mlp_model.predict(X_test_scaled)
test_meta_X = np.column_stack((test_pred_linear, test_pred_lasso, test_pred_mlp))
test_meta_pred = meta_model_opt.predict(test_meta_X)
r2_test, mse_test, rmse_test, mae_test = evaluate_model(y_test, test_meta_pred)

# Tạo DataFrame để hiển thị kết quả đánh giá
results_stac = {
    'Metric': ['R²', 'MSE', 'RMSE', 'MAE'],
    'Train': [r2_train, mse_train, rmse_train, mae_train],
    'Validation': [r2_val, mse_val, rmse_val, mae_val],
    'Test': [r2_test, mse_test, rmse_test, mae_test]
}

results_df2 = pd.DataFrame(results_stac)

# Hiển thị DataFrame
print(results_df2)
```
